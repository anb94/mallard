```python
import pandas as pd
import os
import numpy as np
```

```python
QM_file = "/Users/WCVan/Downloads/pdac_analysis_Collision_data/QM.top.table.tsv"

qm_pda = pd.read_csv(
    QM_file, comment="#", header=0, sep='\t'
)

classical_file = "/Users/WCVan/Downloads/pdac_analysis_Collision_data/classical.top.table.tsv" 

classical = pd.read_csv(
    classical_file, comment="#", header=0, sep='\t'
)

exocrine_file = "/Users/WCVan/Downloads/pdac_analysis_Collision_data/ExocrineLike.top.table.tsv"

exocrine_like = pd.read_csv(
    exocrine_file, comment="#", header=0, sep='\t'
)


human_annot_file = '/Users/WCVan/Downloads/pdac_analysis_Collision_data/homosapiens-annot.txt'

human_annot = pd.read_csv(
    human_annot_file, comment="#", header=0, sep='\t'
)



```

```python
human_annot
```

```python
protein_coding= human_annot[human_annot['GENEBIOTYPE'] == 'protein_coding']
protein_coding
```

```python
qm_pda 
```

```python
qm_pda = qm_pda.loc[:, ['Gene.Symbol', 'adj.P.Val', 'P.Value', 't', 'B', 'logFC']]
qm_pda.rename(columns={'Gene.Symbol': 'ID'}, inplace=True)
qm_pda['ORF'] = qm_pda['ID']
qm_pda
```

```python
qm_pda = qm_pda[qm_pda.ID.isin(protein_coding.GENENAME)]
qm_pda
```

```python
# check duplicated values in the 'ID' column
duplicates = qm_pda[qm_pda.duplicated('ID', keep=False)]
duplicates
```

```python
#aggregate results to handle duplicates
aggre_qm_pda = qm_pda.loc[qm_pda.groupby('ID')['P.Value'].idxmin()].reset_index(drop=True)
aggre_qm_pda
```

```python
# check duplicated values in the 'ID' column
duplicates = aggre_qm_pda[aggre_qm_pda.duplicated('ID', keep=False)]
duplicates
```

```python
qm_pda_sig = aggre_qm_pda[aggre_qm_pda['adj.P.Val'] < 0.05]
qm_pda_sig = qm_pda_sig[qm_pda_sig['logFC'].abs() > 1]
qm_pda_sig
```

```python
qm_pda_neg = aggre_qm_pda[aggre_qm_pda['adj.P.Val'] < 0.05]
qm_pda_neg = qm_pda_neg[qm_pda_neg['logFC'].abs() < 0.5]
qm_pda_neg
```

```python
classical = classical.loc[:, ['Gene.Symbol', 'adj.P.Val', 'P.Value', 't', 'B', 'logFC']]
classical.rename(columns={'Gene.Symbol': 'ID'}, inplace=True)
classical['ORF'] = classical['ID']
classical
```

```python
# check duplicated values in the 'ID' column
duplicates = classical[classical.duplicated('ID', keep=False)]
duplicates
```

```python
#aggregate results to handle duplicates
aggre_classical = classical.loc[classical.groupby('ID')['P.Value'].idxmin()].reset_index(drop=True)
aggre_classical
```

```python
classical_sig = aggre_classical[aggre_classical['adj.P.Val'] < 0.05]
classical_sig = classical_sig[classical_sig['logFC'].abs() > 1]
classical_sig
```

```python
classical_neg = aggre_classical[aggre_classical['adj.P.Val'] < 0.05]
classical_neg = classical_neg[classical_neg['logFC'].abs() < 0.5]
classical_neg
```

```python
exocrine_like = exocrine_like.loc[:, ['Gene.Symbol', 'adj.P.Val', 'P.Value', 't', 'B', 'logFC']]
exocrine_like.rename(columns={'Gene.Symbol': 'ID'}, inplace=True)
exocrine_like['ORF'] = exocrine_like['ID']
exocrine_like
```

```python
#aggregate results to handle duplicates
aggre_exocrine_like = exocrine_like.loc[exocrine_like.groupby('ID')['P.Value'].idxmin()].reset_index(drop=True)
aggre_exocrine_like
```

```python
# check duplicated values in the 'ID' column
duplicates = aggre_exocrine_like[aggre_exocrine_like.duplicated('ID', keep=False)]
duplicates
```

```python
exocrine_sig = aggre_exocrine_like[aggre_exocrine_like['adj.P.Val'] < 0.05]
exocrine_sig = exocrine_sig[exocrine_sig['logFC'].abs() > 1]
exocrine_sig 
```

```python
exocrine_neg = aggre_exocrine_like[aggre_exocrine_like['adj.P.Val'] < 0.05]
exocrine_neg = exocrine_neg[exocrine_neg['logFC'].abs() < 0.2]
exocrine_neg
```
