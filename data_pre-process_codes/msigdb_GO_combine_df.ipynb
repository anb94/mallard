```python
import numpy as np
import pandas as pd
import os
from functools import reduce
```

**ALL Genes**

```python
# get the files
msigdb_all_df = pd.read_csv('C:/Users/WCVan/Downloads/pdac_analysis_vangie/local_basal_results/ML_input_files/ClosenessCentrality_tables/msigdb/local_basal_all.csv')
GO_all_df = pd.read_csv('C:/Users/WCVan/Downloads/pdac_analysis_vangie/local_basal_results/ML_input_files/ClosenessCentrality_tables/GO/local_basal_all.csv')

# Drop the first column "previous row numbers of dataframe"
msigdb_all_df = msigdb_all_df.drop(msigdb_all_df.columns[0], axis=1)
GO_all_df = GO_all_df.drop(GO_all_df.columns[0], axis=1)

# Remove row numbers (index)
msigdb_all_df.reset_index(drop=True, inplace=True)
GO_all_df.reset_index(drop=True, inplace=True)
```

```python
msigdb_all_df
```

```python
GO_all_df
```

```python
# combine dataframes
results_all_df = pd.merge(msigdb_all_df, GO_all_df, how='outer', on='hgnc_symbol')
```

```python
# Sort the genes ID alphabetically
results_all_df = results_all_df.sort_values('hgnc_symbol')
```

```python
results_all_df
```

```python
# check duplicated values in the 'hgnc_symbol' column
duplicates = results_all_df[results_all_df.duplicated('hgnc_symbol', keep=False)]
duplicates
```
