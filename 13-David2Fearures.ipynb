```python
use_terms_df = P_file[P_file['FDR']<0.05].sort_values(by = 'FDR')
use_terms_df

# sorting the list with FDR < 0.05
```

```python
use_category_list = list(use_terms_df['Category'])
use_category_list

# extract the 'Category' column data
```

```python
df = use_terms_df[['Category','Genes']]
df1 = df['Genes'].str.split(', ',expand = True).stack()
df1 

# extract the 'Genes' column data
```

```python
df1 = df1.reset_index(level = 1, drop = True).rename('Genes')
df1

# removes the inner index level and updates the index, and rename the column to 'Genes'
```

```python
df = df.drop('Genes', axis = 1)
new_df=df.join(df1)

# drop the 'Genes' column from df and then join the modified df with df1 to create a new DataFrame new_df
```

```python
new_df
```

```python
grouped = new_df.groupby('Genes')

# creating a new object called 'grouped' by grouping the DataFrame 'new_df' based on the values in the 'Genes' column
```

```python
print (grouped.size().index)
```

```python
forcc = []
for i in grouped.size().index:
    forcc.append(grouped.get_group(i))

# creating an empty list called 'forcc' and then iterating over the groups generated by the grouped object
# The retrieved group is then appended to the forcc list
```

```python
cced = pd.concat(forcc)
cced

# concatenating the list 'forcc', which contains individual groups obtained from the grouped object, into a single DataFrame called 'cced'
```

```python
P_feature = pd.DataFrame()
P_feature['Symbol'] = cced['Genes'].drop_duplicates()
P_feature[[use_category_list]] = 0
P_feature

# creates an empty DataFrame called 'P_feature'
# adds a column named 'Symbol' to the 'P_feature' DataFrame and assigns it the unique values from the 'Genes' column of the 'cced' DataFrame 
# remove duplicate values
# assigns a value of 0 to multiple columns in the 'P_feature' DataFrame as 'use_category_list' is assumed to be a list containing the names of the columns to be updated with 0
```

```python
for temp_gene in cced['Genes']:
    for temp_category in list(cced[cced['Genes'] == temp_gene].Category):
        P_feature.loc[P_feature['Symbol'] == temp_gene,temp_category] = 1

# updating the values in the 'P_feature' DataFrame based on the values in the 'Genes' and 'Category' columns of the 'cced' DataFrame
```
